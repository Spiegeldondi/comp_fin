```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 0.02
sigma = 0.7
T = 1
K = 12
S = 15
```

\begin{align*}
\frac{\partial V}{\partial S_{t}} &= {\mathrm{K}} e^{-r(T-t)} \Phi^{ \prime}(d_2) \, \frac{\partial d_2}{\partial S_t}, \\
\frac{\partial V}{\partial S_{t}} &= {\mathrm{K}} e^{-r(T-t)} \Phi^{ \prime}(d_2)  \, \frac{1}{\sigma \sqrt{T-t}\,S_t}, \quad
\textrm{with:} \quad \Phi^{ \prime}(d_2) = \frac{1}{\sqrt{2 \, \pi}} \, \mathrm{exp}({\frac{-d_2^2}{2}}).
\end{align*}

```python
def d2(S, t):
    return (np.log(S/K) + (r-sigma**2/2)*(T-t)) / (sigma * np.sqrt(T-t))

def phiPrime(d):
    return 1/np.sqrt(2*np.pi)*np.exp(-(d**2)/2)

def delta(S, t):
    return K*np.exp(-r*(T-t))*phiPrime(d2(S, t))*1/(sigma*np.sqrt(T-t))*1/S
```

```python
x = np.linspace(0,0.99,50)
data = delta(S,x)

plt.figure(figsize=(16,9))
plt.plot(x,data)
plt.xlabel("t")
plt.ylabel("Δ") # Δ ∂V/∂S
plt.title("Delta of Put Cash or Nothing")
plt.show()
```
