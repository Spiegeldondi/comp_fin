```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining Equation and Derivative
def f(x):
    res = np.cos(x)-2*x**3
    return res
 
def dfdx(x):
    res = -np.sin(x)-6*x**2
    return res

#### Functions ####

def f1(x):
    return 1/(1+x**2) - 1/2

def f2(x):
    return x**3 - 3*(x**2) - 1

def f3(x):
    return x**3 - 2*x + 2

def f4(x):
    return np.sin(x)

#### Derivatives ####

def df1dx(x):
    return -2*x / (1 + x**2)**2

def df2dx(x):
    return 3*(x-2)*x

def df3dx(x):
    return 3*x**2 - 2

def df4dx(x):
    return np.cos(x)
```

```python
# Newton Method
def newton(f, dfdx, x0):
      if dfdx(x0) == 0:
            print("Division by zero: derivative is 0 at initial guess")
            return x0
      max_iter = 20  # Max iterations
      tol = 1E-15    # Tolerance
      i = 0          # Iteration counter
      xi = x0
      print('Iteration ' + str(i) + ': x = ' + str(x0) + ', f(x) = ' + str(f(x0)))
      # Iterating until either the tolerance or max iterations is met
      while abs(f(xi)) > tol and i < max_iter:
            i = i + 1
            x_new = xi - f(xi) / dfdx(xi)
            print('Iteration ' + str(i) + ': x = ' + str(x_new) + ', f(x) = ' + str(f(x_new)))
            xi = x_new

      return xi
```

```python
def plotNewton(f, dfdx, x0, a, b):
    x_plot = np.linspace(a, b, 800)
    y_plot = f(x_plot)

    xi = newton(f, dfdx, x0)

    fig = plt.figure(dpi=200)
    plt.axhline(y=0, c="black")
    plt.plot(x_plot, y_plot, c='blue')
    # plot computed root
    plt.plot(xi, f(xi), c='red', marker='o', fillstyle='none')
    # indicate initial guess
    plt.axvline(x=x0, c="green")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid()
    plt.show()
```

```python
plotNewton(f1, df1dx, 2, -4, 4)
plotNewton(f2, df2dx, 0, -4, 4)
plotNewton(f3, df3dx, 2, -4, 4)
plotNewton(f4, df4dx, np.arctan(2*np.pi), -4, 140)
```

```python
def bisection(f,a,b):
    if np.sign(f(a)) == np.sign(f(b)):
        print("Zero or even number of roots")
        return f((a+b)/2)
    max_iter = 100   # Max iterations
    tol = 1E-15      # Tolerance
    i = 0            # Iteration counter
    c = (a+b)/2      # new midpoint
    while i < max_iter and (b - a)/2 > tol:
        i += 1
        c = (a+b)/2
        if np.sign(f(a)) == np.sign(f(c)):
            a = c
        else:
            b = c
        if i%10 == 0:
            print('Iteration ' + str(i) + ': c = ' + str(c) + ', f(c) = ' + str(f(c)))
    return c
```

```python
def plotBisection(f, a, b):
    x_plot = np.linspace(-4, 4, 800)
    y_plot = f(x_plot)

    xi = bisection(f, a, b)

    fig = plt.figure(dpi=200)
    plt.axhline(y=0, c="black")
    plt.plot(x_plot, y_plot, c='blue')
    # plot computed root
    plt.plot(xi, f(xi), c='red', marker='o', fillstyle='none')
    # indicate initial guess
    plt.axvline(x=a, c="green")
    plt.axvline(x=b, c="green")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid()
    plt.show()
```

```python
plotBisection(f1, 0, 2)
plotBisection(f2, 0, 4)
plotBisection(f3, -2, 2)
plotBisection(f4, -2, 2.1)
```
