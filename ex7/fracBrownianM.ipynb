```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import cholesky
```

```python
H = 0.5

def fracBrown(steps,a,b):
    t = np.linspace(a, b, steps)
    l = steps
    S = np.zeros((l, l))
    for i, ti in enumerate(t):
        for j, tj in enumerate(t):
            S[i,j] = 0.5*(ti**(2*H) + tj**(2*H) - abs(ti - tj)**(2*H))
    sigma = cholesky(S, lower=True)
    v = np.random.normal(size=l)
    u = sigma@v
    return u
```

```python
u = fracBrown(100,1e-9,1)
plt.figure(dpi=200)
plt.plot(u)
```
