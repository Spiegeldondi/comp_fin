```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
S_n = np.linspace(1, 1000, 100)

# Parameter set for high probability of negative values
kappa = 10
theta = 10

# Feller condition 2*kapppa*theta >= sigma**2
safety_buffer = abs(kappa * theta)
sigma = np.sqrt(2 * kappa * theta) + safety_buffer
```

```python
plt.figure(dpi=200)
plt.xlabel(r"$S_n$")
plt.ylabel(r"$\mathbf{P}[S_{n+1} < 0]$")
plt.title('Parameters: kappa = {}, theta = {}, sigma = {:.2f}'.format(kappa, theta, float(sigma)))

for dt in [0.01, 0.02, 0.05, 0.1, 0.2, 0.5, 1]:
    plt.plot(S_n, norm.cdf(-(S_n + kappa*(theta - S_n)*dt) / (sigma*np.sqrt(S_n*dt))), label = "dt = {}".format(dt))

plt.legend()
```
