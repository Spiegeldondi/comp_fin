```python
import numpy as np
from numpy.random import poisson
import matplotlib.pyplot as plt
```

## Problem 4 

### (ii)

```python
def N(lam, T, n):
    dt = T/n
    process = poisson(lam, n)
    process = np.cumsum(process)
    np.insert(process, 0, 0.0)
    return process

x = np.linspace(0, 1, 100)

lambda1 = 0.1
lambda2 = 0.5

plt.figure(figsize=(16, 10))
plt.scatter(x, N(lambda1, 1, 100))
plt.scatter(x, N(lambda2, 1, 100))
plt.xlabel("t")
plt.ylabel("N(t)")
plt.title("Poisson Process")
```

### (iii)

```python
Tlen = 100
Nt = N(0.5, 1, Tlen)
Nlen = Nt[-1]

# Algorithm script page 53

U = np.random.uniform(size=Nt[-1])
print(U.mean())
Y = np.random.normal(size=Nt[-1])

timesteps = np.linspace(0,1,Tlen)
X = np.zeros(Tlen)

temp = np.zeros(Nlen)
for t,x in enumerate(X):
    for i in range(Nlen): #  Nt[t]
        temp[i] = Y[i]*U[i] if U[i] <= t/Tlen else 0
    X[t] = np.sum(temp)

plt.figure(figsize=(16, 10))
plt.scatter(timesteps, X)
plt.xlabel("t")
plt.ylabel("X(t)")
plt.title("Compound Poisson Process")
```
