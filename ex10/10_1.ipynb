```python
import numpy as np
import matplotlib.pyplot as plt

# Domain and grid width
x = np.linspace(-2, 2, 41)
h = x[1] - x[0]

# Define function and 2nd derivative
u = x**5 + x**4 + x**3 + x**2 + x
uxx = 20*x**3 + 12*x**2 + 6*x + 2

# Initialize solution arrays
uxx_approx_1 = np.zeros((len(x)))
uxx_approx_2 = np.zeros((len(x)))
err1 = []
err2 = []
```

```python
# Approximation 1
for i in range(2, len(x)-2):
    uxx_approx_1[i] = (1/h**2) * (-1/12*u[i+2] + 4/3*u[i+1] - 5/2*u[i] + 4/3*u[i-1] - 1/12*u[i-2]) # O(h^4)
    err1.append(abs(uxx[i]-uxx_approx_1[i]))

# Approximation 2
for i in range(1, len(x)-1):
    uxx_approx_2[i] = (1/h**2) * (u[i + 1] - 2*u[i] + u[i - 1]) # O(h^2)
    err2.append(abs(uxx[i]-uxx_approx_2[i]))
```

```python
# Plot analytic and numeric solution
plt.figure(dpi=200)
plt.plot(uxx[2:-3], label="analytic")
plt.plot(uxx_approx_1[2:-3], marker="o", label="Approximation 1")
plt.plot(uxx_approx_2[2:-3], marker="^", label="Approximation 2")
plt.legend()
plt.grid()
plt.show()

# Plot the error
plt.figure(dpi=200)
plt.loglog(err1, marker="^", label="Approximation 1")
plt.loglog(err2, marker="o", label="Approximation 2")
plt.legend()
plt.grid()
plt.show()
```
