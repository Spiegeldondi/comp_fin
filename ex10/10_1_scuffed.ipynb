```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function u(x)
def u(x):
    return np.sin(x)
```

```python
# Compute the exact second derivative
x = np.linspace(0, 6*np.pi, 1000)
dxx_u = np.cos(x)
```

```python
# Approximation 1
h = 0.1
dxx_u_approx1 = np.zeros(len(x))

for i in range(2, len(x)-2):
    dxx_u_approx1[i] = 1/h**2 * (-1/12 * u(x[i]+2*h) + 4/3 * u(x[i]+h) - 5/2 * u(x[i]) + 4/3 * u(x[i]-h) - 1/12 * u(x[i]-2*h))
```

```python
# Approximation 2
dxx_u_approx2 = np.zeros(len(x))

for i in range(1, len(x)-1):
    dxx_u_approx2[i] = 1/h**2 * (u(x[i]+h) - 2*u(x[i]) + u(x[i]-h))
```

```python
# Compute the error for each approximation
error1 = dxx_u - dxx_u_approx1
error2 = dxx_u - dxx_u_approx2

# Plot the results
plt.plot(x, dxx_u, label='Exact second derivative')
plt.plot(x, dxx_u_approx1, label='Approximation 1')
plt.plot(x, dxx_u_approx2, label='Approximation 2')
plt.xlabel('x')
plt.ylabel('Second derivative of u(x)')
plt.legend()
plt.show()

plt.plot(x, error1, label='Error Approximation 1')
plt.plot(x, error2, label='Error Approximation 2')
plt.xlabel('x')
plt.ylabel('Error')
plt.legend()
plt.show()
```
