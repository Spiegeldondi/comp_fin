```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 10]

ST = np.linspace(0, 100, 101)

K1 = 60
K2 = 40

call_val = np.where(ST-K1 >= 0, ST-K1, 0)
put_val  = np.where(K2-ST >= 0, K2-ST, 0)
```

```python
plt.plot(ST, call_val, lw=4, label="value call option")
plt.plot(ST, put_val, lw=4, label="value put option")
plt.legend(fontsize=24)
```
